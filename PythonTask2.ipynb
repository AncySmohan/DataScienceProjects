1.Write a python program to find those numbers which are divisible by 7 and multiples of 5,for user input

```python
user_input = int(input("Enter a number : "));

if user_input % 35 == 0:
  print(f'The {user_input} is a multiple of 7 and 5')
else:
  print(f'The number {user_input} is not a multiple of either 7 or 5')
```

2.Write a python program to check the user input is even or odd or zero.

```python
user_input = int(input("Enter a number : "));

if user_input % 2 == 0:
  print("{} is an even number".format(user_input))
else:
  print("{} is an odd number".format(user_input))
```

3.Print decimal part of an input number. If the decimal part is zero print the string : “INTEGER”

```python
floatInput = float(input("Enter a decimal number : "))


decimalPart = floatInput % 1

if decimalPart == 0:
  print("INTEGER")
else:
  print(decimalPart)
```
