1.Write a Python program to create a class representing a Circle. Include methods to calculate its area and perimeter.

```python
class Circle:
  def __init__(self,radius):
    self.radius = radius

  def area(self):
    return 3.14 * self.radius * self.radius

  def perimeter(self):
    return 2*3.14*self.radius


small = Circle(3)
print("Area",small.area())
print("perimeter",small.perimeter())

medium = Circle(10)
print("Area",medium.area())
print("perimeter",medium.perimeter())

large = Circle(121)
print("Area",large.area())
print("perimeter",large.perimeter())
```

2.Write a Python program to create a calculator class. Include methods for basic arithmetic operations.

```python
class Calculator:
  def __init__(self):
    pass

  def add(self,a,b):
    return a+b

  def substract(self,a,b):
    return a-b

  def multiply(self,a,b):
    return a*b

  def divide(self,a,b):
    return a/b

  def difference(self,a,b):
    return abs(a-b)

calculator = Calculator()

print("Add",calculator.add(22,45))
print("Substract",calculator.substract(67,10))
print("Multiply",calculator.multiply(22,22))
print("Divide",calculator.divide(121,11))
print("Difference",calculator.difference(99,10))
```
