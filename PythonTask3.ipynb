3.Write a Python program to count the number of strings from a given list of strings. The string length is 2 or more and the first and last characters are the same.
Sample List : ['abc', 'xyz', 'aba', '1221'] Expected Result : 2

```python
l = ['abc', 'xyz', 'aba', '1221']

count = 0

for i in l:
  if len(i) >= 2 and i[-1] == i[0]:
    count += 1

print(count)
```

9.Write a Python program to check if a list is a palindrome or not. Return true otherwise false.

```python
pld = "abcba"

def f(s):
  i = 0
  while i <= (len(s)/2):
    if s[i] == s[-1*(i+1)]:
      i += 1
    else:
      return False
  return True

print(f(pld))
```

10.Write a Python a program to find the union and intersection of two lists.

```python
l1 = [1,2,3,4,5]
l2 = [4,2,8,0]

u = list(set(l1) | set(l2))
i = list(set(l1) & set(l2))

print("Union of both lists is {}".format(u))
print("Intersection of both lists is {}".format(i))
```

14.Write a Python program to create a dictionary with a number and its corresponding square from 1 to input number. And also check if the input number is less than 10 Eg: Input : 3 Output : {1:1, 2:4, 3:9}

```python
Number = int(input('Enter a number : '))

if Number < 0 or Number > 10:
  print('Opperation not supported')
else:
  squares = {i:i*i for i in range(1,1+Number)}
  print(squares)
```

17.Write a Python program to get the top three items in a shop.
Sample data: {'item1': 45.50, 'item2':35, 'item3': 41.30, 'item4':55, 'item5': 24} Expected Output: item4 55 item1 45.5 item3 41.3

```python
allItemsDict =  {'item1': 45.50, 'item2':35, 'item3': 41.30, 'item4':55, 'item5': 24}

sortedByDescByPrice = sorted(allItemsDict.items(),key=lambda x:x[1],reverse=True)

topItemsDict = dict(sortedByDescByPrice[:3])
print(topItemsDict)
```
