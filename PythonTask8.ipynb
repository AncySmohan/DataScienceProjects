##3. Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.
```
Sample Financial data (fdata.csv):
Date,Open,High,Low,Close
10-03-16,774.25,776.065002,769.5,772.559998
10-04-16,776.030029,778.710022,772.890015,776.429993
10-05-16,779.309998,782.070007,775.650024,776.469971
10-06-16,779,780.47998,775.539978,776.859985
10-07-16,779.659973,779.659973,770.75,775.080017
The code snippet gives the output shown in the following screenshot:
```



image.png

````python
import pandas as pd
import matplotlib.pyplot as plt


cols = ['Date','Open','High','Low','Close']
data = [
    ['10-03-16',774.25,776.065002,769.5,772.559998],
    ['10-04-16',776.030029,778.710022,772.890015,776.429993],
    ['10-05-16',779.309998,782.070007,775.650024,776.469971],
    ['10-06-16',779,780.47998,775.539978,776.859985],
    ['10-07-16',779.659973,779.659973,770.75,775.080017],
]


df = pd.DataFrame(data,columns=cols)


plt.plot(df['Date'],df['Open'], label = "Open")
plt.plot(df['Date'],df['High'], label="High")
plt.plot(df['Date'],df['Low'], label = "Low")
plt.plot(df['Date'],df['Close'], label = "Close")

plt.legend()
````

##4.  Write a Python program to plot two or more lines with legends, different widths and colors. The code snippet gives the output shown in the following screenshot:

````python
l1 = [(10,20),(20,40),(30,10)]
l2 = [(10,40),(20,10),(30,30)]

x1,y1 = zip(*l1)
x2,y2 = zip(*l2)

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.plot(x1,y1, linewidth=3, color='blue', label="line1-width-3")
plt.plot(x2,y2, linewidth=5, color='red', label="line2-width-5")

plt.legend()

plt.show()
````

##6.  Write a Python programming to display a bar chart of the popularity of programming Languages.
```
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
The code snippet gives the output shown in the following screenshot:
```

image.png

````python
import pandas as pd
import matplotlib.pyplot as plt


x = ['Java','Python','PHP',"JavaScript",'C#',"C++"]
y = [22.2,17.6,8.8,8,7.8,6.8]

# Create title
plt.title('Popularity of Programming Language/nWorldwide, Oct 2017 compared to a year ago')

# Set the display boundary
plt.axis([-0.5,5.5,0,25])

# Label the axis
plt.xlabel('Languages')
plt.ylabel('Popularity')

# plot the chart
plt.bar(x,y,color='b',edgecolor='black')

# Show the grid
plt.grid(which='major', linestyle='-',color="red")
# turn on the minor grid
plt.minorticks_on()
plt.grid(which='minor', linestyle=':',color="gray")

# Display the plot in screeen
plt.show()
````
